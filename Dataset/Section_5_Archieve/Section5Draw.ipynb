```python
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

def plot_error_vs_time(csv_file_path, save_as_filename):
    """
    Function to plot running time vs error for different proc_count (proc4, proc16, proc64)
    and save the plot as an image, with a stylish theme, optional log scale,
    and display mean labels with exact values.
    
    :param csv_file_path: str, the relative path of the CSV file
    :param save_as_filename: str, the name of the output plot image (e.g., "output.png")
    """
    # Load the CSV data from the provided relative path
    df = pd.read_csv(csv_file_path)
    
    # Filter data for each processor count (proc4, proc16, proc64)
    proc4_df = df[df['proc_count'] == 4]
    proc16_df = df[df['proc_count'] == 16]
    proc64_df = df[df['proc_count'] == 64]
    
    # Compute the mean values for each processor count group
    proc4_mean = proc4_df[['numerical_error', 'running_time']].mean()
    proc16_mean = proc16_df[['numerical_error', 'running_time']].mean()
    proc64_mean = proc64_df[['numerical_error', 'running_time']].mean()

    # Set a stylish theme
    sns.set(style="ticks")  # Dark background with clean ticks and grid
    sns.set_palette("dark")  # Use dark color palette for style
    
    # Create a figure for the plot
    plt.figure(figsize=(8, 6))

    # Plot for proc4 with a circle ('o') marker
    sns.scatterplot(x=proc4_df['numerical_error'], y=proc4_df['running_time'], s=120, 
                    marker='o', label='Proc4', color='dodgerblue')
    
    # Plot for proc16 with a square ('s') marker
    sns.scatterplot(x=proc16_df['numerical_error'], y=proc16_df['running_time'], s=120, 
                    marker='s', label='Proc16', color='forestgreen')
    
    # Plot for proc64 with a triangle ('^') marker
    sns.scatterplot(x=proc64_df['numerical_error'], y=proc64_df['running_time'], s=120, 
                    marker='^', label='Proc64', color='firebrick')

    # Plot the mean points for proc4, proc16, and proc64
    sns.scatterplot(x=[proc4_mean['numerical_error']], y=[proc4_mean['running_time']], 
                    s=200, marker='o', color='dodgerblue', edgecolor='black', linewidth=2, 
                    label=f'Proc4 Mean: ({proc4_mean["numerical_error"]:.3e}, {proc4_mean["running_time"]:.3e})')
    
    sns.scatterplot(x=[proc16_mean['numerical_error']], y=[proc16_mean['running_time']], 
                    s=200, marker='s', color='forestgreen', edgecolor='black', linewidth=2, 
                    label=f'Proc16 Mean: ({proc16_mean["numerical_error"]:.3e}, {proc16_mean["running_time"]:.3e})')
    
    sns.scatterplot(x=[proc64_mean['numerical_error']], y=[proc64_mean['running_time']], 
                    s=200, marker='^', color='firebrick', edgecolor='black', linewidth=2, 
                    label=f'Proc64 Mean: ({proc64_mean["numerical_error"]:.3e}, {proc64_mean["running_time"]:.3e})')

    # Set axis labels
    plt.xlabel('Numerical Error', fontsize=12)
    plt.ylabel('Running Time (s)', fontsize=12)
    
    # Add a title
    plt.title('Running Time vs Numerical Error for Proc4, Proc16, Proc64', fontsize=14)
    
    # Remove the top and right spines for a cleaner look
    sns.despine()

    # Display the legend inside the plot, and remove some clutter to make the plot compact
    plt.legend(title='Processor Count', fontsize=10, loc='upper left', bbox_to_anchor=(0.6, 0.75), frameon=False)
    
    # Save the plot as an image
    plt.savefig(save_as_filename, bbox_inches='tight')
    
    # Show the plot
    plt.show()

# Example usage:
# plot_error_vs_time("relative/path/to/your/file.csv", "comparison_mean_inside_plot.png")
```
